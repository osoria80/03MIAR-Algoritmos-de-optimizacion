```python
from tensorflow import keras
from tensorflow.keras import layers, Model

def build_1d_cnn(input_size=16000, num_classes=10):
    inputs = keras.Input(shape=(input_size, 1))

    # Bloque 1: CL1 + PL1
    x = layers.Conv1D(filters=16, kernel_size=64, strides=2, padding="same")(inputs)
    x = layers.ReLU()(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=8, strides=8)(x)

    # Bloque 2: CL2 + PL2
    x = layers.Conv1D(filters=32, kernel_size=32, strides=2, padding="same")(x)
    x = layers.ReLU()(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(pool_size=8, strides=8)(x)

    #  Bloque 3: CL3
    x = layers.Conv1D(filters=64, kernel_size=16, strides=2, padding="same")(x)
    x = layers.ReLU()(x)
    x = layers.BatchNormalization()(x)

    # Bloque 4: CL4
    x = layers.Conv1D(filters=128, kernel_size=8, strides=2, padding="same")(x)
    x = layers.ReLU()(x)
    x = layers.BatchNormalization()(x)

    # Capas densas: FC1 + FC2 + salida
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.25)(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.25)(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    #Consrtuimos el modelo y lo compilamos
    model = Model(inputs=inputs, outputs=outputs, name="1D_CNN")
    model.compile(
        optimizer=keras.optimizers.Adadelta(learning_rate=1.0),
        loss="mean_squared_logarithmic_error",
        metrics=["accuracy"]
    )
    return model


modelo = build_1d_cnn()
modelo.summary()
```
